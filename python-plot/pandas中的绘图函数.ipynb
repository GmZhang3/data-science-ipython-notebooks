### 1.1 线形图

Series

```python
import pandas as pd
import numpy as np
series = pd.Series(np.random.randn(10).cumsum(),index = np.arange(10))
series.plot()#默认kind = 'line'，线形图
```

dataframe

```python
frame = pd.DataFrame(np.random.randn(10,4).cumsum(axis = 0),index = np.arange(10),columns=['a','b','c','d'])
frame.plot() 
```

### 1.2 柱状图

series

```python
fig,axes = plt.subplots(3,1)
data = pd.Series(np.random.rand(16),index = list('abcdefghijklmnop'))
data.plot(kind = 'bar',ax = axes[0],color = 'k',alpha = 0.7)

data2 = pd.Series(np.random.rand(10),index = list('abcdefghij'))
data2.plot(kind = 'bar',ax = axes[1],color = 'r',alpha = 0.7)

#利用series.value_counts()图形化显示Series中各值得出现频率
data3 = pd.Series(['a','b','c','d','d','a','s','c','c'])
data3.value_counts().plot(kind = 'bar',ax = axes[2],color = 'g',alpha = 0.7)
```

dataframe   
每一行的值为一组

```python
frame = pd.DataFrame(np.random.rand(6,4),index = ['one','two','three','four','five','six'],columns= ['a','b','c','d'])
print frame
frame.plot(kind = 'bar')
```
