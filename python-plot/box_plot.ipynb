# 利用matplotlib画箱线图

```python
#参考：https://blog.csdn.net/You_are_my_dream/article/details/53435733
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1000)
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
data = np.random.normal(size = (1000,),loc = 0,scale = 1) #生成normal (Gaussian) distribution的随机数据，loc方差，scale标准差
data
```

```python

# sym 调整好异常值的点的形状
# whis 默认是1.5， 通过调整它的竖直来设置异常值显示的数量，
# 如果想显示尽可能多的异常值，whis设置很小，否则很大
#四分位距IQR=Q3-Q1，那么上限=Q3+1.5IQR
plt.boxplot(data, sym ="o", whis = 1.5)
#plt.boxplot(data, sym ="o", whis = 0.01)
#plt.boxplot(data, sym ="o", whis = 999)

#--------------------- 
#作者：有一种宿命叫无能为力 
#来源：CSDN 
#原文：https://blog.csdn.net/You_are_my_dream/article/details/53435733 
#版权声明：本文为博主原创文章，转载请附上博文链接！
```

# 同一图中画多个箱型图

```python
labels = ["A","B","C","D","E"]
np.random.seed(100)
data = np.random.normal(loc = 0,scale = 1,size = (1000,5))
data
```

```python
plt.boxplot(data,labels = labels, whis = 1.5,sym = "o")
```
