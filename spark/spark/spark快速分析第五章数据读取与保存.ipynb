```python
#引入并初始化pyspark
import findspark
findspark.init()
```

```python
#python中初始化spark
from pyspark import SparkContext, SparkConf
conf = SparkConf().setMaster("local").setAppName("my app")#集群的URL，应用名
sc = SparkContext(conf = conf)
```

    文本文件读取与保存

```python
#创建一个名为README.MD的RDD，弹性分布式数据集
lines = sc.textFile("README.MD")
```

```python
lines.collect()
```

```python
lines.saveAsTextFile("REDAME_SAVE.MD")
```

    JSON文件读取与保存
    以文本文件的形式读取，使用json解析器对RDD中的值进行操作

```python
import json
json_file = sc.textFile("JSON.MD")
data = json_file.map(lambda x:json.load(x))
type(data)
data.first()
```
